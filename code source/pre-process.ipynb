# 0. Import library

```python
import pandas as pd
import numpy as np
from scipy.io import loadmat
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Import data

```python
mat = loadmat('Database.mat')
mat.keys()
```

```python
df = pd.DataFrame(data=np.array(mat['Datan']), columns=[i[0] for i in mat['nomcol'][0]])
df
```

```python
cycle = pd.read_csv('CycleNumber.txt', header = 0)
date = pd.read_csv('Date.txt', header = 0)
```

```python
df['CYCLE'] = cycle['CYCLE_NUMBER']
df['DATE'] = date['DATE']
df
```

```python
visualize_df = df.drop(['DATE', 'CYCLE'], axis=1)
visualize_df
```

## 2.2. Pairplots

Nous representons chaque variable avec la variable `actual.values`

```python
rows = 12
columns = 2
params = {'axes.titlesize':'30',
          'xtick.labelsize':'10',
          'ytick.labelsize':'10'}
matplotlib.rcParams.update(params)

fig, axs = plt.subplots(rows, columns, figsize=(15,50))
cols = np.reshape(visualize_df.columns[visualize_df.columns != 'actual.values'], (rows,columns))

fig.set

for r in range(rows):
    for c in range(columns):
        axs[r, c].scatter(visualize_df[cols[r][c]], visualize_df['actual.values'])
        axs[r, c].set(xlabel=cols[r][c], ylabel='actual.values')
```

```python
fig.savefig('pairplot.jpg', dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)
```

# 3. Remove useless columns

```python
useless_cols = ['FEED_TMP', 'FEED_DS_030', 'FEED_DS_050', 'FEED_DS_070', 'DIESEL_DS_030', 
                'DIESEL_DS_050', 'DIESEL_DS_070', 'FEED_MeABP']

cols = [c for c in df.columns if c not in useless_cols]
fs_df = df[cols]

cols.remove('CYCLE')
cols.remove('DATE')
cols.remove('PLANT_NAME')

cols
visualize_df = visualize_df[cols]
```

## Correlation matrix after removing

# 4. Remove rows having missing values

```python
params = {'axes.titlesize':'30',
          'xtick.labelsize':'10',
          'ytick.labelsize':'10'}

matplotlib.rcParams.update(params)
plt.figure(figsize=(15,5))
plt.bar(fs_df.isna().sum().index, fs_df.isna().sum())
plt.xticks(rotation=45)
plt.xlabel('Variables', fontsize=15)
plt.ylabel('Nb observations', fontsize=15)
plt.savefig('missing-value.jpg', dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)
```

```python
fs_woNaN_df = fs_df.dropna()
fs_woNaN_df
```

# 5. Remove sites having less than 50 elements

```python
count_site = fs_woNaN_df.groupby('PLANT_NAME').count()
count_site.reset_index(inplace=True)
count_site['PLANT_NAME'] = count_site['PLANT_NAME'].astype(str)

params = {'axes.titlesize':'30',
          'xtick.labelsize':'20',
          'ytick.labelsize':'20'}

matplotlib.rcParams.update(params)


plt.figure(figsize=(20,5))
plt.yticks(np.arange(min(count_site['FEED_D154'])-1, max(count_site['FEED_D154']), 50))
plt.bar(count_site['PLANT_NAME'], count_site['FEED_D154'])
plt.xlabel('Sites', fontsize=25)
plt.ylabel('Fréquences', fontsize=25)

plt.savefig('count-site.jpg', dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None, metadata=None)
```

```python
sites = count_site[count_site['FEED_D154'] >= 50]['PLANT_NAME'].astype(float)
sites
```

```python
fs_woNaN_ss_df = fs_woNaN_df[fs_woNaN_df['PLANT_NAME'].isin(sites)]
fs_woNaN_ss_df
```

# 6. Detect and remove outliers

```python
from scipy import stats
final_df = fs_woNaN_ss_df[(np.abs(stats.zscore(fs_woNaN_ss_df[fs_woNaN_ss_df.columns[:-2]])) < 3).all(axis=1)]
final_df
```

```python
final_df.to_csv('preprocessed-data.csv', index=False)
```
